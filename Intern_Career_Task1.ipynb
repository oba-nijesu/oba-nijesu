```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

### Importing the datasets given

```python
titanictrain = pd.read_csv('train.csv')
titanictrain
```

```python
titanictest1 = pd.read_csv('test.csv')
titanictest1
```

```python
titanictest2 = pd.read_csv('gender_submission.csv')
titanictest2
```

### Data Cleaning Process

```python
#For better test dataset let us merge titanictest1 and titanictest2 datasets together.
titanictest = pd.merge(titanictest1, titanictest2, on='PassengerId', how='inner')
titanictest
```

```python
#From a previously made analysis and machine model making the  accuracy was low due to not enough data to train the model.
#Merging both data
titanic = pd.concat([titanictrain, titanictest], ignore_index=True)
titanic
```

```python
#Bringing out needed columns``
titanic = titanic[['Pclass','Age','SibSp','Parch','Fare','Survived']]
titanic
```

### Removing Nan values

```python
titanic = titanic.dropna()
titanic
```

```python
over_20 = titanic[(titanic['Age'] >= 18) & (titanic['Age'] <= 23)]
over_20
```

```python
titanic['SibSp']
```

```python
titanic['Age'].plot.hist(color = 'black');
plt.title('Distribution of Ages on the Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency');
```

#### From the plot we are able to deduce the highest age range of people are of 18-23

```python
titanic.plot.hist(figsize = (10,30), subplots = True, bins = 100);
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
fig,(ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10,10),
                              sharex = True)

#Add data to ax0
scatter = ax0.scatter(x = over_20['Age'],
                      y = over_20['Pclass'],
                      c = over_20['Survived'],
                      cmap = 'winter')

#Customize ax0
ax0.set(title = "Titanic Survival Due to Class",
        ylabel = "Pclass")

#Add a legend
ax0.legend(*scatter.legend_elements(), title = 'Survived')

#Add a mean line
ax0.axhline(y = over_20["Pclass"].mean(),
           linestyle = '--')



#Add data to ax1
scatter = ax1.scatter(x = over_20['Age'],
                      y = over_20['Fare'],
                      c = over_20['Survived'],
                      cmap = 'winter')

#Customize ax0
ax1.set(title = "Titanic Survival Due to Fare Price",
       xlabel = "Age",
       ylabel = "Fare Price");

#Add a legend
ax1.legend(*scatter.legend_elements(), title = 'Survived');

#Add a mean line
ax1.axhline(y = over_20["Fare"].mean(),
           linestyle = '--')

#Add a Title to the figure
fig.suptitle("Titanic Survival(18-23) Analysis",
             fontsize = 17,
           fontweight = 'bold');
```

#### From the first plot we can see that from the age 18-23 more of the people of this age range on first class survived and no one on second class survived and 22% of thirdclass survived.

#### From the second plot people of age 22 with the fare range of 15-75 survived more than the rest

```python
x = titanic[['Pclass','Age','SibSp','Parch','Fare']]
y = titanic['Survived']
```

```python
x
```

```python
y
```

```python
#2. Choose the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

#We will keep the default hyperparameters
clf.get_params()
```

```python
#fit the model to the training data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
clf.fit(x_train, y_train)
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
clf.score(x_test, y_test)
```

```python
#Apart from acuracy there are other evaluation metrics
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_preds))
```

```python
#Improve the model above
#We are going to try different types of n_estimators

best_accuracy = 0
best_estimators = 0

np.random.seed(42)
for i in range(10,70,10):
    print(f'Trying model with {i} estimators...')
    clf = RandomForestClassifier(n_estimators = i).fit(x_train,y_train)
    accuracy = clf.score(x_test, y_test) * 100
    print(f'Model accuracy on test set: {accuracy:.2f}%')
    print(' ')

if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_estimators = i
print(f'The best model is with {best_estimators} estimators, with an accuracy of {best_accuracy:.2f}%')
```

```python
#Save the model and load it.
import pickle
pickle.dump(clf, open('random_forst_model_1.pk1', 'wb'))
```

```python
loaded_model = pickle.load(open('random_forst_model_1.pk1', 'rb'))
loaded_model.score(x_test, y_test)
```

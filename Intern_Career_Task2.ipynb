```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
bc = breast_cancer.dropna(axis = 1)
bc
```

## Visualization for Fractal Dimension

```python
import matplotlib.pyplot as plt

viz = bc[['fractal_dimension_mean', 'fractal_dimension_se', 'fractal_dimension_worst']]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))  # Create subplots for each histogram

# Plot each histogram with a different color
colors = ['orange', 'green', 'purple']
for i, (col, color) in enumerate(zip(viz.columns, colors)):
    viz[col].hist(ax=axes[i], color=color, bins=50)
    axes[i].set_title(col)

plt.tight_layout(pad=3.0);
```

## Data Preprocessing

## Changing M & B to binary values

##### where M = 1 and B = 0

```python
bc['diagnosis'] = np.where(bc['diagnosis'] == 'M', 1, 0)
bc
```

```python
bc
```

```python
x = bc[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean','radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se', 'fractal_dimension_se', 'radius_worst', 'texture_worst', 'perimeter_worst', 'area_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst', 'fractal_dimension_worst']]
x
```

```python
y = bc['diagnosis']
y
```

## Model Building

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.get_params()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
clf.fit(x_train, y_train)
```

## Model Evaluation

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
clf.score(x_test, y_test)
```

## Model Tuning

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
#Improving the model above
#We are going to try different types of n_estimators
np.random.seed(42)
for i in range(10,100,10):
    print(f'Trying model with {i} estimators...')
    clf = RandomForestClassifier(n_estimators = i).fit(x_train,y_train)
    print(f'Model accuracy on test set : {clf.score(x_test,y_test) * 100: .2f}%')
    print(' ')
```

```python
import pickle
pickle.dump(clf, open('Breast_Cancer_Model.pk1', 'wb'))
```

```python
loaded_model = pickle.load(open('Breast_Cancer_Model.pk1', 'rb'))
loaded_model.score(x_test, y_test)
```

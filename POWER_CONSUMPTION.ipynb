```python
pc1 = pc[['Datetime','Temperature','Humidity','WindSpeed','GeneralDiffuseFlows','DiffuseFlows','PowerConsumption_Zone1']]
pc1
```

```python
pc1['PowerConsumption_Zone1'].describe()
```

```python
pc1['PowerConsumption_Level'] = pc1['PowerConsumption_Zone1'].apply(lambda x: 1 if x > 25000 else 0)
pc1
```

```python
h.plot(kind = 'scatter',
                x = 'Temperature',
                y = 'PowerConsumption_Zone1',
                c = 'PowerConsumption_Level');
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

clf.get_params()
```

```python
b = pc1.drop('PowerConsumption_Zone1',axis = 1)
b
```

```python
c = b.drop('PowerConsumption_Level',axis =1)
```

```python
x = c.drop('Datetime',axis = 1)
x
```

```python
y = pc1['PowerConsumption_Level']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
clf.fit(x_train, y_train)
```

```python
y_preds = clf.predict(x_test)
y_preds
```

```python
y_test
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
np.random.seed(42)
for i in range(10,100,10):
    print(f'Trying model with {i} estimators...')
    clf = RandomForestClassifier(n_estimators = i).fit(x_train,y_train)
    print(f'Model accuracy on test set : {clf.score(x_test,y_test) * 100: .2f}%')
    print(' ')
```
